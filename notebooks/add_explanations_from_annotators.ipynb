```python
target_string = "explain the reasoning behind your answer"
for part in range(1, 3):
    print(part)
    for particip_id in range(1, 11, 1):
        print('v'+str(particip_id))
        df = pd.read_csv(rf"C:\Users\Nursulu_1\Downloads\ContraDetect\automatic_stance_detection\data\qualtrics_survey\responses_part_{part}\V{particip_id}.csv")
        # Find columns where row[0] in the **next** column contains the target string
        selected_cols = []
        for i in range(len(df.columns) - 1):  # Stop before the last column
            current_col = df.columns[i]
            next_col = df.columns[i + 1]
            
            # Check if row[0] in the next column contains the target string
            if isinstance(df.iloc[0, i + 1], str) and target_string.lower() in df.iloc[0, i + 1].lower():
                selected_cols.append(current_col)  # Save current column
                selected_cols.append(next_col)  # Save next column

        # Remove duplicates (if any)
        selected_cols = list(dict.fromkeys(selected_cols))  

        # Filter DataFrame
        filtered_df = df[selected_cols]
        
        annotator = f'v{particip_id}'
        for column in filtered_df.columns:
            string = filtered_df[column].loc[0]
            if re.match(r"^Q\d+", string):
                txt1, txt2 = filtered_df[column].loc[0].split("\n\n")[1], filtered_df[column].loc[0].split("\n\n")[2]
                sample = txt1 + "\n\n" + txt2
                label = filtered_df[column].iloc[-1]
                
            elif target_string in string:
                explanation = filtered_df[column].iloc[-1]
        
                try:
                    sample_to_explanations[sample]['explanations'].append(explanation) 
                    sample_to_explanations[sample]['annotators'].append(annotator) 
                    sample_to_explanations[sample]['labels'].append(label) 

                except:
                    sample_to_explanations[sample] = dict()
                    sample_to_explanations[sample]['explanations'] = [explanation]
                    sample_to_explanations[sample]['annotators'] = [annotator]
                    sample_to_explanations[sample]['labels'] = [label]
            
    
            
```

```python
df_gold = pd.read_csv(r"C:\Users\Nursulu_1\Downloads\ContraDetect\automatic_stance_detection\data\qualtrics_survey\results\df_gold_all.csv")
```

```python
sample_to_explanations_majority_agrees = dict()
```

```python
for el in sample_to_explanations:
    # print(el)
    # go through samples
    gold_label = df_gold[df_gold['Input'] == el]['Answer'].iloc[0]
    for annotator, annot_answer, explanation in zip(sample_to_explanations[el]['annotators'], sample_to_explanations[el]['labels'], sample_to_explanations[el]['explanations']):
        if gold_label == annot_answer:
            try:
                sample_to_explanations_majority_agrees[el]['explanation'].append(explanation)
                sample_to_explanations_majority_agrees[el]['annotator'].append(annotator)
            except:
                sample_to_explanations_majority_agrees[el] = dict() 
                sample_to_explanations_majority_agrees[el]['explanation'] = [explanation]
                sample_to_explanations_majority_agrees[el]['annotator'] = [annotator]
                
        
```
